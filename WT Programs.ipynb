### 23. Write an Ajax program to get book details from an XML file when the user selects a book name. Create an XML file for storing details of a book(title, author, year, price).

```python
#xml file
<?xml version="1.0" encoding="UTF-8"?>
<books>
    <book>
        <title>Java</title>
        <author>James Gosling</author>
        <year>1995</year>
        <price>500</price>
    </book>

    <book>
        <title>Python</title>
        <author>Guido van Rossum</author>
        <year>1991</year>
        <price>400</price>
    </book>

    <book>
        <title>C++</title>
        <author>Bjarne Stroustrup</author>
        <year>1985</year>
        <price>450</price>
    </book>
</books>
```

```python
#php file
<!DOCTYPE html>
<html>
<head>
    <title>Book Details using AJAX</title>

    <script>
        function getBookDetails() {
            var selectedBook = document.getElementById("book").value;

            var xhttp = new XMLHttpRequest();
            xhttp.onreadystatechange = function () {
                if (this.readyState == 4 && this.status == 200) {
                    displayDetails(this, selectedBook);
                }
            };

            xhttp.open("GET", "books.xml", true);
            xhttp.send();
        }

        function displayDetails(xml, bookName) {
            var xmlDoc = xml.responseXML;
            var books = xmlDoc.getElementsByTagName("book");
            var output = "";

            for (var i = 0; i < books.length; i++) {
                var title = books[i].getElementsByTagName("title")[0].childNodes[0].nodeValue;

                if (title == bookName) {
                    var author = books[i].getElementsByTagName("author")[0].childNodes[0].nodeValue;
                    var year = books[i].getElementsByTagName("year")[0].childNodes[0].nodeValue;
                    var price = books[i].getElementsByTagName("price")[0].childNodes[0].nodeValue;

                    output = "<b>Title:</b> " + title + "<br>" +
                             "<b>Author:</b> " + author + "<br>" +
                             "<b>Year:</b> " + year + "<br>" +
                             "<b>Price:</b> " + price;
                }
            }

            document.getElementById("result").innerHTML = output;
        }
    </script>
</head>

<body>

    <h2>Select Book</h2>

    <select id="book" onchange="getBookDetails()">
        <option value="">--Select Book--</option>
        <option value="Java">Java</option>
        <option value="Python">Python</option>
        <option value="C++">C++</option>
    </select>

    <div id="result" style="margin-top:20px;"></div>

</body>
</html>
```
